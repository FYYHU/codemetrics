```python
import pandas as pd
```

```python
code_metric_df = pd.read_csv("codemetrics.csv", sep=";")
simian_dupe_df = pd.read_csv("simian_duplicates.csv")
```

```python
simian_dupe_df
```

```python
# Extract second part of the directory
simian_dupe_df['file_space'] =simian_dupe_df['file_path'].apply(lambda x: x.split('/')[1] if len(x.split('/')) > 1 else None)
```

```python
simian_dupe_df
```

```python
file_space_group_df = simian_dupe_df.groupby('file_space').sum(numeric_only=True)
```

```python
file_space_group_df
```

```python
cleaned_code_metric_df = code_metric_df[code_metric_df["Scope"]=="Assembly"]
cleaned_code_metric_df["Project"] = cleaned_code_metric_df["Project"].str.replace(r'\s*\(Debug\)', '', regex=True)
```

```python
cleaned_code_metric_df
```

```python
print(cleaned_code_metric_df.drop(columns=["Namespace","Type", "Member", "Scope"]).to_latex(index=False))
```

```python
temp = file_space_group_df.reset_index().merge(cleaned_code_metric_df, left_on="file_space", right_on="Project")
cleanup = temp.drop(columns=["Type", "Member", "Namespace", "start_line", "end_line"])
```

```python
cleanup
```

```python
print(cleanup.drop(columns=["Scope","Project"]).to_latex(index=False))
```

```python
cleanup.set_index("file_space").corr(numeric_only=True)
```
